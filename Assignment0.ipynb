### Colour channels are being read in the wrong order. To rectify this we need to use cv2.cvtColor

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
img2 = cv2.imread("img2.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.imread("img3.jpeg")
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img4 = cv2.imread("img4.jpg")
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
```

```python
plt.subplot(2,2,1)
plt.imshow(img1)
plt.subplot(2,2,2)
plt.imshow(img2)
plt.subplot(2,2,3)
plt.imshow(img3)
plt.subplot(2,2,4)
plt.imshow(img4)
```

### Converting image 3 to grayscale and saving in images folder
#### Weights for RGB while converting to grayscale were obtained empirically (by researchers) by observing luminence perception wrt colour
#### Formula used: Linear approximation of gamma correction

```python
#img3_gray = cv2.imread("img3.jpeg", cv2.IMREAD_GRAYSCALE)

l, b, c = img3.shape
img3_gray = np.zeros((l, b))

for i in range(l):
    for j in range(b):
        img3_gray[i][j] = (0.299*img3[i][j][0] + 0.587*img3[i][j][1] + 0.114*img3[i][j][2])
        
plt.imshow(img3_gray, cmap="gray")
plt.axis("off")
plt.savefig("../images/img3gray.jpeg")
```
